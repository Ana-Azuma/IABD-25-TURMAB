```python
# importando as bibliotecas 
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates # biblioteca matplotlib que permite formatação do gráfico no eixo x
```

```python
# cria variavel df para armazenar o dataframe
df = pd.read_csv('https://raw.githubusercontent.com/danielvieira95/Inteligencia_Artificial_2025/refs/heads/main/bases_de_dados/monitoramento_tempo.csv')
```

```python
df.head() # exibi sas 4 primeiras linhas do dataframe
```

```python
df.tail() # exibe as 4 ultimas linhas do dataframe
```

```python
# importar a biblioteca datetime para fazer conversao dos dados de data de objeto para datetime
import datetime 
df['data'] = pd.to_datetime(df['data']) # conversão da coluna data para datetime 64
```

```python
df.info() # exibe informacoes sobre o dataframe
```

```python
plt.plot(df['data'], df['temperatura']) # plota o grafico temperatura x data
```

```python
# criando codigo para plotar um grafico dentro do utro 
fig = plt.figure(figsize=(12,5)) # cria o tamanho da figura
eixo = fig.add_axes([0,0,1,1]) # cria os eixos da figura
eixo.set_ylim(200,500)  # define os limites do grafico
eixo.plot(df['data'], df['temperatura'], color='b') # plota o grafico
eixo.grid(True) # ativa a grade do grafico
eixo.set_xlim(datetime.datetime(2014,5,1), datetime.datetime(2014,6,1)) # define o periodo do grafico
eixo.set_title('Temperatura no momento',fontsize=25,pad=20) # titulo do grafico
eixo.set_xlabel('Ano',fontsize=20) # legenda do eixo x
eixo.set_ylabel('Temperatura',fontsize=20) # legenda do eixo y
plt.legend() # exibe a legenda do grafico


# criando x 2 dentro do grafico
eixo2=fig.add_axes([0.35, 0.50,0.5,0.3]) # cria o segundo eixo para o grafico

# definindo intervalo do eixo x
eixo2.xaxis.set_major_locator(mdates.DayLocator(interval=7)) 
eixo2.xaxis.set_major_formatter(mdates.DateFormatter('%d-%b')) 

# Defininfo periodo do grafico desejado
eixo2.set_xlim(datetime.datetime(2015,4,1), datetime.datetime(2015,5,1)) # define o periodo do grafico desejado
eixo2.tick_params(axis='x', labelsize=8) # espaçemnto do eixo x
eixo2.plot(df['data'], df['temperatura'], color='r') # plotando o grafico de temperatura
eixo2.tick_params(axis='x', labelsize=8) # espaçemnto do eixo x
eixo.grid(True) # ativa a grade do grafico
eixo2.set_title('Temperatura x ano',fontsize=15) # titulo do grafico
eixo2.set_ylabel('Temperatura',fontsize=20) # legenda do eixo y
eixo2.set_xlabel('Ano',fontsize=20) # legenda do eixo x
eixo2.legend(['Temperatura'],loc='best',fontsize=10) # exibe a legenda do grafico
```

```python
fig=plt.figure(figsize=(15,8)) # cria o tamanho da figura
plt.rcParams['font.size']='10' # fonte da figura
eixo = fig.add_axes([0,0,1,1]) # cria o eixo 1
eixo.set_ylim(200,500)  # define os limites do eixo y
eixo2 = fig.add_axes([0.65,0.58,0.3,0.3]) # cria o eixo 2
eixo.plot(df['data'], df['temperatura'], color='b') # plota o grafico
eixo.grid(True) # ativa a grade do grafico

# define o limite do grafico
eixo.set_xlim(datetime.datetime(2014,5,1), datetime.datetime(2014,6,1)) # define o periodo do grafico
eixo.set_title('Temperatura no momento',fontsize=20) # titulo do grafico
eixo.set_xlabel('Ano',fontsize=20) # legenda do eixo x
eixo.set_ylabel('Temperatura',fontsize=20) # legenda do eixo y
plt.legend(['temperatura'],loc='best') # exibe a legenda do grafico
azul_esquerda=df['data']<datetime.datetime(2014,5,1) # armazena o periodo
azul_direita=df['data']>datetime.datetime(2014,6,1) # armazena o periodo

# eixo 2
eixo2.set_xlim(datetime.datetime(2014,1,1), datetime.datetime(2015,1,1))
eixo2.plot(df['data'], df['temperatura'], color='r') 
eixo2.plot(df['data'][azul_esquerda], df['temperatura'][azul_esquerda], color='b')
eixo2.plot(df['data'][azul_direita], df['temperatura'][azul_direita], color='b')
eixo2.grid(True)
eixo2.set_title('Temperatura - 05/2014',fontsize=15) 
eixo2.set_ylabel('Temperatura',fontsize=15) 
eixo2.set_xlabel('Mês',fontsize=15)
eixo2.legend(['Temperatura'],loc='best',fontsize=10) 

```
