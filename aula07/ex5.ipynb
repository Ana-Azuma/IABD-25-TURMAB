```python
# Importando bibliotecas
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Dados de entrada
consumo = np.array([50, 140, 30, 75, 123]).reshape(-1, 1)
demanda = np.array([100, 200, 80, 100, 160])

# Criando o modelo de regress√£o linear
modelo = LinearRegression()
modelo.fit(consumo, demanda)

# Prevendo a demanda para o mesmo conjunto de consumo
demanda_prevista = modelo.predict(consumo)

# Exemplo: prever a demanda para um consumo de 90KW
consumo_novo = np.array([[90]])
demanda_nova = modelo.predict(consumo_novo)
print(f"Demanda prevista para 90KW: {demanda_nova[0]:.2f} KW")
```

```python
# Plotando os pontos reais e a linha de regress√£o
plt.scatter(consumo, demanda, color='blue', label='Dados reais')
plt.plot(consumo, demanda_prevista, color='red', label='Linha de regress√£o')
plt.scatter(consumo_novo, demanda_nova, color='green', label='Previs√£o 90KW', s=100)

plt.xlabel('Consumo (KW)')
plt.ylabel('Demanda (KW)')
plt.title('Previs√£o de Demanda de Energia')
plt.legend()
plt.grid(True)
plt.show()
```
