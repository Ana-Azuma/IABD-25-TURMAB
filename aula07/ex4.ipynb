```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
# Dados fornecidos
Renda = np.array([1200, 5000, 8000, 10000, 30000]).reshape(-1, 1)
Gastos = np.array([750, 1500, 4000, 6000, 18000])

# Modelo 1: Regressão Linear Simples
linear_model = LinearRegression()
linear_model.fit(Renda, Gastos)
Gastos_linear_pred = linear_model.predict(Renda)

# Modelo 2: Regressão Polinomial (grau 2)
poly = PolynomialFeatures(degree=2)
Renda_poly = poly.fit_transform(Renda)
poly_model = LinearRegression()
poly_model.fit(Renda_poly, Gastos)

# Geração de valores para curvas
Renda_range = np.linspace(1000, 31000, 500).reshape(-1, 1)
Gastos_linear_curve = linear_model.predict(Renda_range)
Gastos_poly_curve = poly_model.predict(poly.transform(Renda_range))
```

```python
# Plotagem
plt.figure(figsize=(10, 6))
plt.scatter(Renda, Gastos, color='black', label='Dados Reais')
plt.plot(Renda_range, Gastos_linear_curve, color='green', label='Regressão Linear Simples')
plt.plot(Renda_range, Gastos_poly_curve, color='blue', linestyle='--', label='Regressão Polinomial (grau 2)')
plt.xlabel('Renda')
plt.ylabel('Gastos')
plt.title('Previsão de Gastos com Base na Renda')
plt.legend()
plt.grid(True)
plt.show()
```
