```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
# Dados de treinamento
A = np.array([50, 100, 120, 300]).reshape(-1, 1)  # Área em m²
Valor = np.array([180000, 300000, 375000, 600000]).reshape(-1, 1)  # Valor em R$

# Criar e treinar o modelo de regressão linear
modelo = LinearRegression()
modelo.fit(A, Valor)

# Dados de teste (novas áreas para prever o valor)
A_teste = np.array([60, 150, 200]).reshape(-1, 1)  # Exemplo de áreas novas
Valor_previsto = modelo.predict(A_teste)

# Imprimir previsões
for i in range(len(A_teste)):
    print(f"Área: {A_teste[i][0]} m² - Valor previsto: R$ {Valor_previsto[i][0]:,.2f}")
```

```python
# Plotar o gráfico
plt.scatter(A, Valor, color='blue', label='Dados de Treinamento')
plt.scatter(A_teste, Valor_previsto, color='red', label='Dados de Teste')
plt.plot(A, modelo.predict(A), color='green', label='Regressão Linear')
plt.xlabel('Área (m²)')
plt.ylabel('Valor (R$)')
plt.title('Relação entre Área e Valor do Imóvel')
plt.legend()
plt.grid(True)
plt.show()
```
