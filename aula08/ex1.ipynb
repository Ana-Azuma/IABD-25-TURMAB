```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dados = {
    'Velocidade de Operação': [10, 5, 8, 6, 12, 4],
    'Complexidade da Tarefa': ['Baixa', 'Alta', 'Média', 'Alta', 'Baixa', 'Alta'],
    'Manutenção Necessária': ['Baixa', 'Alta', 'Média', 'Alta', 'Baixa', 'Média'],
    'Classificação': ['Montagem', 'Teste', 'Montagem', 'Teste', 'Montagem', 'Teste']
}

df = pd.DataFrame(dados)
print("Dataset:")
print(df)
print()
```

```python
le_complexidade = LabelEncoder()
le_manutencao = LabelEncoder()
le_classificacao = LabelEncoder()
```

```python
X = df.copy()
X['Complexidade da Tarefa'] = le_complexidade.fit_transform(df['Complexidade da Tarefa'])
X['Manutenção Necessária'] = le_manutencao.fit_transform(df['Manutenção Necessária'])
X = X[['Velocidade de Operação', 'Complexidade da Tarefa', 'Manutenção Necessária']]

y = le_classificacao.fit_transform(df['Classificação'])
```

```python
arvore = DecisionTreeClassifier(random_state=42, max_depth=3)
arvore.fit(X, y)

print(f"Acurácia: {arvore.score(X, y):.2%}\n")
```

```python
plt.figure(figsize=(15, 10))
tree.plot_tree(arvore, 
               feature_names=['Velocidade', 'Complexidade', 'Manutenção'],
               class_names=['Montagem', 'Teste'],
               filled=True,
               rounded=True,
               fontsize=10)
plt.title('Árvore de Decisão - Máquinas de Montagem vs Teste', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
