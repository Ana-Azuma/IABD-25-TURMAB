```python
# importando as bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import datetime # biblioteca que permite a transformação da data que era um object em datatime 
import matplotlib.dates as mdates # biblioteca matplotlib que permite formatação do gráfico no eixo x
```

```python
# realizar a leitura do meu arquivo csv
df = pd.read_csv('basededados/monitoramento_tempo.csv')
```

```python
# exibindo as primeiras linhas do dataframe
df.head() # exibindo as 4 primeiras linhas do dataframe
```

```python
df['data'] = pd.to_datetime(df['data']) # converção da data de object para datatime
fig=plt.figure(figsize=(15,8))
plt.rcParams['font.size']='20'
eixo=fig.add_axes([0,0,1,1])
eixo.plot(df['data'],df['humidade'],color='b',lw=0.5,linestyle='--')
eixo.set_ylim(0,100)
eixo.set_xlim(datetime.datetime(2014,5,1), datetime.datetime(2014,5,30)) # xlim define o valor inicial e o valor final
eixo.set_title('Umidade - Ano 2015',fontsize=25)
eixo.legend(['Umidade'],fontsize=20)
eixo.set_ylabel('Umidade',fontsize=20)
eixo.set_xlabel('Ano',fontsize=20)
plt.xticks(rotation=90,fontsize=15,ha='right') # ajusta o espaçamento
plt.show()
```

```python
df['data'] = pd.to_datetime(df['data']) # converção da data de object para datatime
fig=plt.figure(figsize=(15,8))
plt.rcParams['font.size']='20'
eixo=fig.add_axes([0,0,1,1])
eixo.plot(df['data'],df['velocidade do vento'],color='b',lw=0.5,linestyle='--')
eixo.set_ylim(0,10)
eixo.set_xlim(datetime.datetime(2014,5,1), datetime.datetime(2014,5,30)) # xlim define o valor inicial e o valor final
eixo.set_title('Velocidade do vento - Ano 2015',fontsize=25)
eixo.legend(['Velocidade do vento'],fontsize=20)
eixo.set_ylabel('Velocidade do vento',fontsize=20)
eixo.set_xlabel('Ano',fontsize=20)
plt.xticks(rotation=90,fontsize=15,ha='right') # ajusta o espaçamento
plt.show()
```

```python
fig = plt.figure(figsize=(12,5)) # cria o tamanho da figura
eixo = fig.add_axes([0,0,1,1]) # cria os eixos da figura
eixo.set_ylim(200,500)  # define os limites do grafico
eixo.plot(df['data'], df['direção do vento'], color='b') # plota o grafico
eixo.grid(True) # ativa a grade do grafico
eixo.set_xlim(datetime.datetime(2014,5,1), datetime.datetime(2014,6,1)) # define o periodo do grafico
eixo.set_title('Direção do vento',fontsize=25,pad=20) # titulo do grafico
eixo.set_xlabel('Ano',fontsize=20) # legenda do eixo x
eixo.set_ylabel('Direção do vento',fontsize=20) # legenda do eixo y
plt.legend() # exibe a legenda do grafico


# criando x 2 dentro do grafico
eixo2=fig.add_axes([0.35, 0.50,0.5,0.3]) # cria o segundo eixo para o grafico

# definindo intervalo do eixo x
eixo2.xaxis.set_major_locator(mdates.DayLocator(interval=5)) 
eixo2.xaxis.set_major_formatter(mdates.DateFormatter('%d-%b')) 

# Defininfo periodo do grafico desejado
eixo2.set_xlim(datetime.datetime(2015,4,1), datetime.datetime(2015,5,1)) # define o periodo do grafico desejado
eixo2.tick_params(axis='x', labelsize=8) # espaçemnto do eixo x
eixo2.plot(df['data'], df['direção do vento'], color='r') 
eixo2.tick_params(axis='x', labelsize=8) # espaçemnto do eixo x
eixo.grid(True) # ativa a grade do grafico
eixo2.set_title('Direção do vento x ano',fontsize=15) # titulo do grafico
eixo2.set_ylabel('Direção do vento',fontsize=20) # legenda do eixo y
eixo2.set_xlabel('Ano',fontsize=20) # legenda do eixo x
eixo2.legend(['Direção do vento'],loc='best',fontsize=10) # exibe a legenda do grafico
```
